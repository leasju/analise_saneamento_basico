#####▫️ Remoção de outliers da coluna 'perc_banagua' para plotar o histograma
Para remoção dos outliers foi necessário:
- Descobrir o valor do 1° e 3° quartil
- Realizar o cálculo do IQR (Intervalo Interquartil): Q3 - Q1
- Calcular os limites superior e inferior
- Remover os outliers

```python
# Limites do IQR para a coluna perc_lixo
Q1 = saneamentoDF["perc_lixo"].quantile(0.25) # Valor do quartil 1
Q3 = saneamentoDF["perc_lixo"].quantile(0.75) # Valor do quartil 3
IQR = Q3 - Q1 # Cálculo do IQR (Intervalo Interquartil)

# Cálculo dos limites (inferior e superior)
limiteInferior = Q1 - 1.5 * IQR
limiteSuperior = Q3 + 1.5 * IQR

# Remoção dos outliers
sanLixoDF = saneamentoDF[
    (saneamentoDF["perc_lixo"] >= limiteInferior) &
    (saneamentoDF["perc_lixo"] <= limiteSuperior)
]
```

```python
print(f"Limite inferior: {limiteInferior} \nLimite superior: {limiteSuperior}")

# Encontrar o índice do município com o valor mais próximo do LI
idx_li = (saneamentoDF["perc_lixo"] - limiteInferior).abs().idxmin()
municipio_li = saneamentoDF.loc[idx_li, "nome_meso"]
valor_li = saneamentoDF.loc[idx_li, "perc_lixo"]

# Encontrar o índice do município com o valor mais próximo do LS
idx_ls = (saneamentoDF["perc_lixo"] - limiteSuperior).abs().idxmin()
municipio_ls = saneamentoDF.loc[idx_ls, "nome_meso"]
valor_ls = saneamentoDF.loc[idx_ls, "perc_lixo"]

print("\n")

# Exibir os resultados
print(f"Município mais próximo de LI ({limiteInferior}): {municipio_li}")
print(f"Município mais próximo de LS ({limiteSuperior}): {municipio_ls}")
```

#### Remoção de outliers e cálculo do limite superior(ls) e limite inferior(li) para buscar qual o melhor município e pior município

Tratamento de Outlier

*   O idxmin() foi utilizado para identificar o menor valor dentro do Limite Inferior
*   O idxmax() foi utilizado para encontrar o maior valor dentro do Limite Superior.
*   Esses códigos foram pesquisados e retirados da documentação da biblioteca Pandas.

```python
# Limites do IQR para a coluna perc_lixo
Q1N = df_norte["perc_lixo"].quantile(0.25)  # Quartil 1
Q2N = df_norte["perc_lixo"].quantile(0.50)  # Quartil 2
Q3N = df_norte["perc_lixo"].quantile(0.75)  # Quartil 3
IQRN = Q3N - Q1N  # Intervalo Interquartil

print(Q1N, Q2N, Q3N, IQRN)

# Cálculo dos limites
limiteInferiorNorte = round(Q1N - 1.5 * IQRN)
limiteSuperiorNorte = round(Q3N + 1.5 * IQRN)

# Remoção de outliers
df_norteMinas = df_norte[
    (df_norte["perc_lixo"] >= limiteInferiorNorte) &
    (df_norte["perc_lixo"] <= limiteSuperiorNorte)
]

#Encontrando a melhor cidade (maior valor de perc_lixo)
melhor_cidadeNorte = df_norteMinas.loc[df_norteMinas["perc_lixo"].idxmax()]
#Encontrando a pior cidade (menor valor de perc_lixo)
pior_cidadeNorte = df_norteMinas.loc[df_norteMinas["perc_lixo"].idxmin()]
```

Análise do melhor município e pior município

- Melhor município: maior taxa de disposição correta do lixo coletado
- Pior município: menor taxa de disposição correta do lixo coletado


```python
# Exibindo os limites para verificar
print("Limite Inferior:", limiteInferiorNorte)
print("Limite Superior:", limiteSuperiorNorte)

print("\n")

# Exibindo os municípios
print("Melhor município:", melhor_cidadeNorte['mun'], "e tem", melhor_cidadeNorte['perc_lixo'], "% de descarte de lixo adequado.")
print("Pior município:", pior_cidadeNorte['mun'], "e tem apenas", pior_cidadeNorte['perc_lixo'], "% de descarte de lixo adequado.")
```

- Descobrir o valor do 1° e 3° quartil
- Realizar o cálculo do IQR (Intervalo Interquartil): Q3 - Q1
- Calcular os limites superior e inferior
- Remover os outliers

```python
# Limites do IQR (Intervalo Interquartil)  para a coluna 'perc_banagua'
Q1 = saneamentoDF["perc_banagua"].quantile(0.25) # Valor do quartil 1
Q3 = saneamentoDF["perc_banagua"].quantile(0.75) # Valor do quartil 3
IQR = Q3 - Q1 # Cálculo do IQR (Intervalo Interquartil)

# Cálculo dos limites (inferior e superior)
limiteInferior = Q1 - 1.5 * IQR
limiteSuperior = Q3 + 1.5 * IQR

# Remoção dos outliers
sanAguaDF = saneamentoDF[
    (saneamentoDF["perc_banagua"] >= limiteInferior) &
    (saneamentoDF["perc_banagua"] <= limiteSuperior)
]
```

```python
print(f"Limite Inferior: {limiteInferior}\nLimite Superior: {limiteSuperior}\n")

# Encontrar o índice do município com o valor mais próximo do LI
idx_li = (saneamentoDF["perc_banagua"] - limiteInferior).abs().idxmin()
municipio_li = saneamentoDF.loc[idx_li, "nome_meso"]
valor_li = saneamentoDF.loc[idx_li, "perc_banagua"]

# Encontrar o índice do município com o valor mais próximo do LS
idx_ls = (saneamentoDF["perc_banagua"] - limiteSuperior).abs().idxmin()
municipio_ls = saneamentoDF.loc[idx_ls, "nome_meso"]
valor_ls = saneamentoDF.loc[idx_ls, "perc_banagua"]

# Exibir os resultados
print(f"Município mais próximo de LI ({limiteInferior}): {municipio_li}, {valor_li}")
print(f"Município mais próximo de LS ({limiteSuperior}): {municipio_ls}, {valor_ls}")
```

#### Remoção de outliers e cálculo do limite superior(ls) e limite inferior(li) para buscar qual o melhor município e pior município

Tratamento de Outlier

*   O idxmin() foi utilizado para identificar o menor valor dentro do Limite Inferior
*   O idxmax() foi utilizado para encontrar o maior valor dentro do Limite Superior.
*   Esses códigos foram pesquisados e retirados da documentação da biblioteca Pandas.

```python
# Limites do IQR para a coluna perc_esgoto
Q1O = df_oeste["perc_esgoto"].quantile(0.25)  # Quartil 1
Q3O = df_oeste["perc_esgoto"].quantile(0.75)  # Quartil 3
IQRO = Q3O - Q1O  #(Intervalo Interquartil)

# Cálculo dos limites
limiteInferiorOeste = round(Q1O - 1.5 * IQRO,2)
limiteSuperiorOeste = round(Q3O + 1.5 * IQRO,2)

# Remoção de outliers
df_oesteMinas = df_oeste[
    (df_oeste["perc_esgoto"] >= limiteInferiorOeste) &
    (df_oeste["perc_esgoto"] <= limiteSuperiorOeste)
]

# Encontrando a melhor cidade (maior valor de perc_esgoto)
melhor_cidadeOeste = df_oesteMinas.loc[df_oesteMinas["perc_esgoto"].idxmax()]
# Encontrando a pior cidade (menor valor de perc_esgoto)
pior_cidadeOeste = df_oesteMinas.loc[df_oesteMinas["perc_esgoto"].idxmin()]
```

- Pior município: menor taxa de água encanada
